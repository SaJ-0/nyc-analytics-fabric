```python
import requests
import os

url = "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2024-02.parquet"
dest = "/lakehouse/default/Files/taxi/yellow_tripdata_2024-02.parquet"

print("скачивание февраля")

response = requests.get(url, stream=True)
print(f"HTTP статус: {response.status_code}")

if response.status_code == 200:
    with open(dest, 'wb') as f:
        for chunk in response.iter_content(chunk_size=8*1024*1024):  # 8MB кусками
            f.write(chunk)
    size = os.path.getsize(dest)
    print(f"Размер: {size / 1024 / 1024:.1f} MB")
else:
    print(f"Ошибка: {response.status_code}")
```

```python
import requests
import os

base_path = "/lakehouse/default/Files/taxi/"
base_url = "https://d37ci6vzurychx.cloudfront.net/trip-data/"

for month in range(1, 13):
    filename = f"yellow_tripdata_2024-{month:02d}.parquet"
    dest = base_path + filename
    
    if os.path.exists(dest):
        size = os.path.getsize(dest) / 1024 / 1024
        print(f"Уже есть: {filename} ({size:.0f} MB)")
        continue
    
    url = base_url + filename
    print(f"Скачиваю: {filename} ...")
    
    response = requests.get(url, stream=True)
    if response.status_code == 200:
        with open(dest, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8*1024*1024):
                f.write(chunk)
        size = os.path.getsize(dest) / 1024 / 1024
        print(f"окей{filename} — {size:.0f} MB")
    else:
        print(f"Ошибка {response.status_code}: {filename}")

print("\nГотово")
```

```python
import requests, json, os, time

API_KEY = "ba5ecdb2f944a38dac5300a1b34dda3e6f4c7190573c7f2b24c450393dd55432"
headers = {"X-API-Key": API_KEY}
base_path = "/lakehouse/default/Files/openaq/"

# Берём первые 10 станций (остальные аналогично, API имеет лимиты)
location_ids = [384, 386, 625, 626, 628, 631, 642, 648, 662, 664]

all_measurements = []

for loc_id in location_ids:
    url = f"https://api.openaq.org/v3/locations/{loc_id}/sensors"
    resp = requests.get(url, headers=headers)
    
    if resp.status_code != 200:
        print(f" Локация {loc_id}: ошибка {resp.status_code}")
        continue
    
    sensors = resp.json().get("results", [])
    print(f"Локация {loc_id}: {len(sensors)} сенсоров")
    
    for sensor in sensors:
        sensor_id = sensor["id"]
        parameter = sensor.get("parameter", {}).get("name", "unknown")
        
        # Берём измерения за 2024 год
        meas_url = f"https://api.openaq.org/v3/sensors/{sensor_id}/measurements/daily"
        meas_params = {
            "date_from": "2024-01-01",
            "date_to": "2024-12-31",
            "limit": 365
        }
        meas_resp = requests.get(meas_url, headers=headers, params=meas_params)
        
        if meas_resp.status_code == 200:
            measurements = meas_resp.json().get("results", [])
            for m in measurements:
                m["location_id"] = loc_id
                m["parameter"] = parameter
            all_measurements.extend(measurements)
        
        time.sleep(0.2)  # не превысить лимит API
    
    print(f"  Всего записей: {len(all_measurements)}")

# Сохраняем всё
with open(base_path + "measurements_2024.json", "w") as f:
    json.dump(all_measurements, f)

print(f"\n Сохранено {len(all_measurements)} измерений")
```
